```matlab
testname='3D_Poisson';

matlab_TT = load(sprintf('plot_data/TT_%s.mat',testname));

matlab_TT_NEs = matlab_TT.NEs;

matlab_TT_err = cellfun(@(c) c.error, matlab_TT.R);

matlab_TT_time = cellfun(@(c) c.time, matlab_TT.R);

matlab_TT_comp = cellfun(@(c) c.Agttcomp, matlab_TT.R);
```

```matlab
fortran_err = [3.9406417333855016E-003,1.0774087203136923E-003,2.8065187946109068E-004,7.1467115927311355E-005,1.7938647201618507E-005];

fortran_time = [9.7748000000000002E-002,0.43733399999999995,2.4482409999999999,10.252182000000001,85.065738999999994];

fortran_comp = [5.1879882812500000E-004,3.2424926757812500E-005,2.0265579223632812E-006,1.2665987014770508E-007,7.9162418842315674E-009];

```

```matlab
plot(matlab_TT_NEs,matlab_TT_err,'-','DisplayName','TT Matlab','color','red');
hold on;
plot(matlab_TT_NEs,fortran_err,'--o','DisplayName','TT Thor Fortran','color','blue');

legend();

xlabel(sprintf('N_E per dimension', Interpreter='latex'));

ylabel(sprintf('Relative Numerical Error',Interpreter='latex'));

set(gca, 'YScale','log');

clf;
```

```matlab
plot(matlab_TT_NEs,matlab_TT_comp,'-','DisplayName','TT Matlab','color','red');
hold on;
plot(matlab_TT_NEs,fortran_comp,'--o','DisplayName','TT Thor Fortran','color','blue');

legend('location','north east');

xlabel(sprintf('N_E per dimension', Interpreter='latex'));

ylabel(sprintf('Compression Ratio',Interpreter='latex'))

set(gca, 'YScale','log');

clf;
```

```matlab
plot(matlab_TT_NEs,matlab_TT_time,'-o','DisplayName','TT Matlab','color','red');
hold on;
plot(matlab_TT_NEs,fortran_time,'--o','DisplayName','TT Thor Fortran','color','blue');

legend('location','north west');

xlabel(sprintf('N_E per dimension', Interpreter='latex'));

ylabel(sprintf('Elapsed Time [Seconds]',Interpreter='latex'))

set(gca, 'YScale','log');

clf;
```
