```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
norder = range(1,11)

matlab_err = np.loadtxt('matlab_plot_data/d100_ncell2.dat')

fortran_err = []

for fn in norder:
    fortran_err.append(   np.loadtxt('fortran_plot_data/output_'+str(fn)+'.dat')      )

fortran_err = np.array(fortran_err)

plt.title('d=100  ncell=2')

plt.plot(norder,matlab_err,label='MATLAB Amen Cross 2',color='C0' )
plt.plot(norder,matlab_err,'o',color='C0' )
plt.plot(norder,fortran_err,label='THOR DMRG' ,color='C1' )
plt.plot(norder,fortran_err,'o' ,color='C1' )

plt.xlabel('Number of nodes')
plt.ylabel('Approximation Error')

plt.yscale('log')

plt.legend()

plt.savefig('plots/f8_d100_ncell2.pdf')
```

```python
matlab_time = np.loadtxt('matlab_plot_data/d100_ncell2_time.dat')
matlab_time_cross = np.loadtxt('matlab_plot_data/d100_ncell2_time_cross.dat')

fortran_time = []
fortran_time_cross = []

for fn in norder:
    fortran_time.append(   np.loadtxt('fortran_plot_data/output_'+str(fn)+'_time.dat')      )
    fortran_time_cross.append(   np.loadtxt('fortran_plot_data/output_'+str(fn)+'_time_cross.dat')      )

fortran_time = np.array(fortran_time)
fortran_time_cross = np.array(fortran_time_cross)

plt.title('d=100  ncell=2')

plt.plot(norder,matlab_time,label='MATLAB Amen Cross 2',color='C0' )
plt.plot(norder,matlab_time,'o',color='C0' )
plt.plot(norder,fortran_time,label='THOR DMRG' ,color='C1' )
plt.plot(norder,fortran_time,'o' ,color='C1' )

plt.ylim(top=10,bottom=0.4)

plt.xlabel('Number of nodes')
plt.ylabel('Time [seconds]')

plt.yscale('log')

plt.legend()

plt.savefig('plots/f8_d100_ncell2_time.pdf')
```

```python
matlab_size = np.loadtxt('matlab_plot_data/d100_ncell2_size.dat')
log_matlab_size = np.log(matlab_size)

fortran_size = []

for fn in norder:
    fortran_size.append(   np.loadtxt('fortran_plot_data/output_'+str(fn)+'_size.dat')      )

fortran_size = np.array(fortran_size)
log_fortran_size = np.log(fortran_size)

log_full_size = 100 * np.log(2*np.linspace(1,10,num=10))

log_fortran_compression = log_fortran_size - log_full_size
log_matlab_compression = log_matlab_size - log_full_size


plt.title('d=100  ncell=2')

plt.plot(norder,np.exp(log_matlab_compression),label='MATLAB Amen Cross 2',color='C0' )
plt.plot(norder,np.exp(log_matlab_compression),'o',color='C0' )
plt.plot(norder,np.exp(log_fortran_compression),label='THOR DMRG' ,color='C1' )
plt.plot(norder,np.exp(log_fortran_compression),'o' ,color='C1' )

plt.xlabel('Number of nodes')
plt.ylabel('Compression')

plt.yscale('log')

plt.legend()

plt.savefig('plots/f8_d100_ncell2_compression.pdf')
```

```python
matlab_max_ranks = np.loadtxt('matlab_plot_data/d100_ncell2_max_ranks.dat')
matlab_avg_ranks = np.loadtxt('matlab_plot_data/d100_ncell2_avg_ranks.dat')


fortran_avg_ranks = []
fortran_max_ranks = []

for fn in norder:
    fortran_avg_ranks.append(   np.loadtxt('fortran_plot_data/output_'+str(fn)+'_avg_rank.dat')      )
    fortran_max_ranks.append(   np.loadtxt('fortran_plot_data/output_'+str(fn)+'_max_rank.dat')      )

fortran_avg_ranks = np.array(fortran_avg_ranks)
fortran_max_ranks = np.array(fortran_max_ranks)

plt.title('d=100  ncell=2')

plt.plot(norder,matlab_max_ranks,label='MATLAB Amen Cross 2 max rank',color='C0' )
plt.plot(norder,matlab_max_ranks,'o',color='C0' )

plt.plot(norder,matlab_avg_ranks,'-.',label='MATLAB Amen Cross 2 average rank',color='C0' )
plt.plot(norder,matlab_avg_ranks,'o',color='C0' )

plt.plot(norder,fortran_max_ranks,label='THOR DMRG max rank',color='C1' )
plt.plot(norder,fortran_max_ranks,'o',color='C1' )

plt.plot(norder,fortran_avg_ranks,'-.',label='THOR DMRG average rank',color='C1' )
plt.plot(norder,fortran_avg_ranks,'o',color='C1' )


plt.xlabel('Number of nodes')
plt.ylabel('Rank')

#plt.yscale('log')

plt.legend()

plt.savefig('plots/f8_d100_ncell2_ranks.pdf')
```
