```matlab
clf

ft_time = load("ft/benchmark-results/time.txt");
tt_time = load("tt/benchmark-results/time.txt");

num_lvls = 4;
Nx = 5*(2.^linspace(1,num_lvls,num_lvls));

plot(Nx,ft_time./tt_time(1:4),'-o');

ylim([0 0.25])
xlim([5 90])

xlabel("$N_x$","interpreter","latex");
ylabel("Speed-up")
```

```matlab
clf

legname   = {"FT","TT"};

ft_l2 = load("ft/benchmark-results/L2.txt");
tt_l2 = load("tt/benchmark-results/L2.txt");

num_lvls = 4;
Nx = 5*(2.^linspace(1,num_lvls,num_lvls));

plot(Nx,ft_l2(1,1:4),'-o');
hold on;
plot(Nx,tt_l2(1,1:4),'-o');

set(gca, 'YScale', 'log')
xlabel("$N_x$","interpreter","latex");
ylabel("$L_2$ Error","interpreter","latex");

legend(legname);
```
