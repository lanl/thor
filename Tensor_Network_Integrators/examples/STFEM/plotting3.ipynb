```matlab
addpath(genpath('../../../matlab/STFEM/src/'))
addpath(genpath('../../../matlab/utils/chebfun/'))
addpath(genpath('../../../matlab/utils/tt-toolbox/'))
addpath(genpath('../../../matlab/utils/ttfunc/'))
```

```matlab
testname='3D_CDR';

matlab_TT = load(sprintf('plot_data/TT_%s.mat',testname));

matlab_TT_NEs = matlab_TT.NEs;

matlab_TT_err = cellfun(@(c) c.error, matlab_TT.R);

matlab_TT_time = cellfun(@(c) c.time, matlab_TT.R);

matlab_TT_comp = cellfun(@(c) c.Agttcomp, matlab_TT.R);
```

```matlab
fortran_err = [6.8136743593678967E-002,2.0042943453308298E-002,5.3893423689134173E-003,1.3938610787751046E-003,3.5417694985191109E-004];

fortran_comp = [4.1198730468750000E-004,6.4373016357421875E-006,1.0058283805847168E-007,1.5716068446636200E-009,2.4556356947869062E-011];

fortran_time = [0.11014199999999999,0.17235899999999998,0.47078799999999998,2.0508599999999997,9.5492770000000000];
```

```matlab
plot(matlab_TT_NEs, matlab_TT_err, '-', ...
    'DisplayName','TT Matlab', ...
    'Color','red', ...
    'LineWidth',2);

hold on;

plot(matlab_TT_NEs, fortran_err, '--o', ...
    'DisplayName','TT THOR Fortran', ...
    'Color','blue', ...
    'LineWidth',2, ...   % thicker line
    'MarkerSize',8, ...  % bigger markers
    'MarkerFaceColor','blue'); % filled markers

    
legend('FontSize',20,'Interpreter','latex');

xlabel(sprintf('N$_E$ per dimension'), 'Interpreter','latex', 'FontSize',20);

ylabel(sprintf('Relative Numerical Error'), 'Interpreter','latex', 'FontSize',20);

set(gca, 'YScale','log', ...
    'FontSize',18, ...
    'LineWidth',1.2, ...
    'TickLabelInterpreter','latex');

axis tight;

exportgraphics(gca, 'figures/Error_comparison.pdf', 'ContentType','vector', 'BackgroundColor','none');

clf;
```

```matlab
plot(matlab_TT_NEs, matlab_TT_comp, '-', ...
    'DisplayName','TT Matlab', ...
    'Color','red', ...
    'LineWidth',2);

hold on;

plot(matlab_TT_NEs, fortran_comp, '--o', ...
    'DisplayName','TT THOR Fortran', ...
    'Color','blue', ...
    'LineWidth',2, ...   % thicker line
    'MarkerSize',8, ...  % bigger markers
    'MarkerFaceColor','blue'); % filled markers

legend('location','north east','FontSize',20,'Interpreter','latex');

xlabel(sprintf('N$_E$ per dimension'), 'Interpreter','latex', 'FontSize',20);

ylabel(sprintf('Compression Ratio'), 'Interpreter','latex', 'FontSize',20);

set(gca, 'YScale','log', ...
    'FontSize',18, ...
    'LineWidth',1.2, ...
    'TickLabelInterpreter','latex');

axis tight;

exportgraphics(gca, 'figures/Compression_comparison.pdf', 'ContentType','vector', 'BackgroundColor','none');

clf;
```

```matlab
plot(matlab_TT_NEs, matlab_TT_time, '-', ...
    'DisplayName','TT Matlab', ...
    'Color','red', ...
    'LineWidth',2);

hold on;

plot(matlab_TT_NEs, fortran_time, '--o', ...
    'DisplayName','TT THOR Fortran', ...
    'Color','blue', ...
    'LineWidth',2, ...   % thicker line
    'MarkerSize',8, ...  % bigger markers
    'MarkerFaceColor','blue'); % filled markers

legend('location','north west','FontSize',20,'Interpreter','latex');

xlabel(sprintf('N$_E$ per dimension'), 'Interpreter','latex', 'FontSize',20);

ylabel(sprintf('Elapsed Time [Seconds]'), 'Interpreter','latex', 'FontSize',20);


set(gca, 'YScale','log', ...
    'FontSize',18, ...
    'LineWidth',1.2, ...
    'TickLabelInterpreter','latex');

axis tight;

exportgraphics(gca, 'figures/Time_comparison.pdf', 'ContentType','vector', 'BackgroundColor','none');

clf;
```
